```python
def split_genres(genres):
    if type(genres) == str:
      return genres.split(", ")
    else:
      return None
```

```python
all_genres = []
for genre in pd.Series(pd.read_csv('/content/drive/MyDrive/finalAIProject/movies_initial.csv')['genre'].unique()):
  if type(genre) == str:
    for i in genre.split(", "):
      if i not in all_genres:
        all_genres.append(i)
  else:
    continue
```
